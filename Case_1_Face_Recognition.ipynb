#Test similarity on my images

```python
# 1. load two faces using load_image_file
face_file1 = "/content/content/SAM/2024_students/Estefani_Santiago_Gatica.jpg"
face_file2 = "/content/content/SAM/2024_students/Estefani_Santiago_Gatica_aged.jpg"
image1 = face_recognition.load_image_file(face_file1)
image2 = face_recognition.load_image_file(face_file2)
```

```python
# 2. Compare the face_encodings similarity using face_distance
img1_encoding = face_recognition.face_encodings(image1)[0]
img2_encoding = face_recognition.face_encodings(image2)[0]

# Calculate the distance between the faces
face_distance = face_recognition.face_distance([img1_encoding], img2_encoding)

# Normalize the distance to get the similarity score
similarity = 1 - face_distance[0]

# Print the similarity score
print("The two faces are '{:.2f}%' similar".format(similarity*100))
```

```python
# 3. is it a match?
face_matches = face_recognition.compare_faces([img1_encoding], img2_encoding)

# Print the result
if face_matches[0]:
    print("The two faces match")
else:
    print("The two faces do not match")
```

```python
dir_name = 'content'
def list_jpeg_files(directory):
    """List all JPEG files in the specified directory."""
    jpeg_files = [file for file in os.listdir(directory) if file.lower().endswith(('.jpeg', '.jpg'))]
    return jpeg_files


# Replace 'your_directory_path' with the path to the directory you want to search
directory_path = '/content/content/SAM/2024_students'
jpeg_files = list_jpeg_files(directory_path)

print("JPEG files in the directory:", jpeg_files)
```

#Encode all faces and print out faces that could not be encoded

```python
import face_recognition

face_dict = {}

for file in jpeg_files:
    # Load image
    img = face_recognition.load_image_file(f"/content/content/SAM/2024_students/{file}")

    # Get face encodings if faces are found
    face_encodings = face_recognition.face_encodings(img)

    if len(face_encodings) > 0:
        # If at least one face is found, take the first face's encoding
        enc = face_encodings[0]
        face_dict[file] = enc
    else:
        # Handle case when no face is found in the image
        face_dict[file] = 0
        print(f"No face found in {file}")
```

```python
# convert to dataframe
data_items = list(face_dict.items())

# Create DataFrame
face_df = pd.DataFrame(data_items, columns=['image', 'encoding'])
face_df.head()
```

```python
def match_faces(image1, image2, face_df):
    # Fetch encodings for the given images from the DataFrame
    enc1_query = face_df.loc[face_df['image'] == image1, 'encoding']
    enc2_query = face_df.loc[face_df['image'] == image2, 'encoding']

    if enc1_query.empty or enc2_query.empty:
        similarity = 0
        print(f"The two faces are {similarity * 100:.2f}% similar")

        return 0
          # Set similarity to 0 if images not found

    enc1 = enc1_query.values[0]
    enc2 = enc2_query.values[0]

    # Calculate the distance between the faces
    face_distance = np.linalg.norm(np.array(enc1) - np.array(enc2))

    # Calculate the similarity score
    similarity = 1 - face_distance

    # Print the similarity score
    print(f"The two faces are {similarity * 100:.2f}% similar")
    return similarity * 100

match_faces('Estefani_Santiago_Gatica.jpg', 'Estefani_Santiago_Gatica_aged.jpg', face_df)
```

```python
match_file['match_similarity'] = match_file.apply(lambda row: match_faces(row['source_image'], row['target_image'], face_df), axis=1)
```

#Create a predicted mathc column to hold if the images match or not with a threshold of 50%

```python
match_file['predicted_match'] = np.where(match_file['match_similarity'] > 50, 'match','no match')
```

#Create confusion matrix and print Accuracy, Precision, and Recall

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

# Calculate confusion matrix
matrix = confusion_matrix(match_file['expected_match'],
                          match_file['predicted_match'],
                          labels=["match", "no match"])

print(matrix)
```

```python
# Calculate confusion matrix
y_true = match_file['expected_match']
y_pred = match_file['predicted_match']
labels = ["match", "no match"]
matrix = confusion_matrix(y_true, y_pred, labels=labels)

# Calculate Accuracy, Precision and Recall
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, pos_label="match", average =None)
recall = recall_score(y_true, y_pred, pos_label="match", average=None)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

```python
# Plotting using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(matrix, annot=True, fmt="d", xticklabels=labels, yticklabels=labels, cmap="Blues")
plt.title("Confusion Matrix")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
